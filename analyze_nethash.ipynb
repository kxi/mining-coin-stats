```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
COIN_BLOCK_INFO_PATH = "coin_nethash_info"

coin = "EXVO"
# coin = "RVN"
# coin = "PGN"
# coin = "LUX"
# coin = "PROTON"
# coin = "RDEN"

fname = os.path.join(COIN_BLOCK_INFO_PATH, coin+"_nethash_info.csv")
```

```python
df = pd.read_csv(fname)
```

```python
print("Number of Nethash Sample: {}".format(df.count()[0]))
```

```python
print("Latest Block: {}".format(int(df["height"][0])))
```

```python
df.head(5)
```

```python
MA_window = min(20, int((df.count()[0]-1)*0.5))
print("MA Windows = {}".format(MA_window))
```

```python
df_ma = df.rolling(window=MA_window).mean()
df_ma.head(10)
```

```python
df.head(10)
```

```python
fig = plt.figure(figsize=(20,10))
x_axis = df['height']
y_axis = df['nethash']
y_axis_ma = df_ma['nethash']

plt.plot(x_axis, y_axis)
plt.plot(x_axis, y_axis_ma)

# plt.plot(x_axis, y_axis)
# fig.suptitle('Block_Speed vs Block Height', fontsize=20)
# plt.xlabel('Block Height', fontsize=18)
# plt.ylabel('Block Speed', fontsize=18)
```

```python
nethash_latest = int(df['nethash'][0])
nethash_mv = int(df_ma['nethash'].head((1)))
```

```python
print("Coin [{}] Recent Nethash is {}".format(coin, nethash_latest))
print("Coin [{}] Nethash MA is {}".format(coin, nethash_mv))
```
